```python
# Extract the 'city' column from the merged dataset
city_column = merged_df['city']

# Plotting the histogram
plt.figure(figsize=(12, 8))  # Adjust the figure size if needed
city_column.value_counts().plot(kind='bar')
plt.title('Histogram of Matches by City')
plt.xlabel('City')
plt.ylabel('Frequency')
plt.xticks(rotation=90)  # Rotate x-labels for better readability if there are many cities
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Extract the 'city' column from the merged dataset
year_column = merged_df['season']

# Plotting the histogram
plt.figure(figsize=(12, 8))  # Adjust the figure size if needed
year_column.value_counts().plot(kind='bar')
plt.title('Histogram of Matches by City')
plt.xlabel('year')
plt.ylabel('Frequency')
plt.xticks(rotation=90)  # Rotate x-labels for better readability if there are many cities
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Maximum wins by Mumbai Indians in 2011 are
wins = data2[(data2['season'] == 2011) & (data2['winner'] == 'Mumbai Indians')]

wins_2011 = wins['winner'].count()

print('Maximum wins by Mumbai Indian in 2011 is :',wins_2011)
```

```python
sns.countplot(x = 'toss_decision',data=data2)
```

```python
# Calculate the counts of each type of toss decision
toss_decision_counts = data2['toss_decision'].value_counts()

# Calculate the percentage of 'bat' and 'field'
bat_percentage = (toss_decision_counts['bat'] / toss_decision_counts.sum()) * 100
field_percentage = (toss_decision_counts['field'] / toss_decision_counts.sum()) * 100

print(f"Percentage of 'bat' in toss_decision: {bat_percentage:.2f}%")
print(f"Percentage of 'field' in toss_decision: {field_percentage:.2f}%")

     
```

```python
# Assuming 'data2' is the DataFrame containing match information
# Counting matches won by teams choosing to bat first after winning the toss
bat_first_wins = data2[(data2['toss_winner'] == data2['winner']) & (data2['toss_decision'] == 'bat')]

# Counting matches won by teams choosing to field first after winning the toss
field_first_wins = data2[(data2['toss_winner'] == data2['winner']) & (data2['toss_decision'] == 'field')]

# Comparing counts
if len(bat_first_wins) > len(field_first_wins):
    print("Teams winning tosses and electing bat first have won more number of matches. Statement B is true.")
elif len(bat_first_wins) < len(field_first_wins):
    print("Teams winning tosses and electing field first have won more number of matches. Statement A is true.")
else:
    print("The number of matches won after choosing bat first and field first are equal.")
```

```python
# Assuming 'data2' is the DataFrame containing match information
# Filtering matches where teams batting second have won
batting_second_wins = data2[data2['win_by_wickets'] > 0]

# Counting the number of matches won by teams batting second for each year
batting_second_wins_by_year = batting_second_wins.groupby('season')['win_by_wickets'].count()

# Finding the year with the highest number of matches won by batting second
top_year = batting_second_wins_by_year.idxmax()

# Finding the second year with the highest number of matches won by batting second
second_top_year = batting_second_wins_by_year.drop(top_year).idxmax()

print("Years where batting second has won more matches:", top_year, second_top_year)
```

```python
# In 2019 which batsman scored highest number of runs by hitting 6's and 4's?
batsman_4s = data1[data1['batsman_runs'] == 4].groupby(by = ['batsman']).agg(fours=pd.NamedAgg(column="batsman_runs", aggfunc="value_counts")).nlargest(n = 10, columns = 'fours').droplevel('batsman_runs')
batsman_4s.head()
```

```python
batsman_6s = data1[data1['batsman_runs'] == 6].groupby(by = ['batsman']).agg(sixs=pd.NamedAgg(column="batsman_runs", aggfunc="value_counts")).nlargest(n = 10, columns = 'sixs').droplevel('batsman_runs')
batsman_6s.head()
```

```python
sns.barplot(x = batsman_4s.index, y = batsman_4s.fours, )
sns.despine()
plt.title('Top 10 Batsmen - Highest number of 4s hit')
```

```python
sns.barplot(x = batsman_6s.index, y = batsman_6s.sixs, )
sns.despine()
plt.title('Top 10 Batsmen - Highest number of 6s hit')
```

```python
# Assuming 'matches_data' is the DataFrame containing match information
# Filtering data for Virat Kohli's performances in 2016
kohli_2016 = merged_df[(merged_df['batsman'] == 'V Kohli') & (merged_df['season'] == 2016)]

# Calculating the total runs scored by Kohli in 2016
total_runs_kohli_2016 = kohli_2016['batsman_runs'].sum()

# Calculating the total balls faced by Kohli in 2016
total_balls_faced_kohli_2016 = len(kohli_2016)

# Calculating the strike rate of Kohli in 2016
strike_rate_kohli_2016 = (total_runs_kohli_2016 / total_balls_faced_kohli_2016) * 100

print(f"The strike rate of Virat Kohli in 2016 is: {strike_rate_kohli_2016:.2f}")

     
```

```python
# Assuming 'matches_data' is the DataFrame containing match information
# Counting the number of matches hosted at each venue
venue_matches_count = data2['venue'].value_counts()

# Finding the venue that has hosted the most matches
most_matches_venue = venue_matches_count.idxmax()
matches_count = venue_matches_count.max()

print(f"The venue that has hosted the most number of IPL matches is '{most_matches_venue}' with {matches_count} matches.")

     
```

```python
# Filter data for run outs
df_filtered = data1[data1['dismissal_kind'] == 'run out']

# Count run outs for each batsman
run_outs = df_filtered['batsman'].value_counts()

# Sort by run outs
run_outs = run_outs.sort_values(ascending=False)

# Get top 3 batsmen with most run outs
top_three_batsmen = run_outs.index[:3].tolist()

# Print the top 3 batsmen with most run outs
print(f"The top 3 batsmen with most run outs in IPL from 2008 to 2019 are: {top_three_batsmen}")
     
```

```python
# Assuming 'matches_data' is the DataFrame containing match information
# Filtering data for deliveries where V Kohli faced JJ Bumrah
kohli_vs_bumrah = data1[(data1['batsman'] == 'V Kohli') & (data1['bowler'] == 'JJ Bumrah')]

# Calculating the total runs scored by Kohli against Bumrah
total_runs_kohli_vs_bumrah = kohli_vs_bumrah['batsman_runs'].sum()

print(f"The total runs scored by V Kohli when facing JJ Bumrah is: {total_runs_kohli_vs_bumrah}")

     
```

```python
# Assuming 'matches_data' is the DataFrame containing match information
# Filtering data for dismissals via caught and bowled
caught_and_bowled_dismissals = data1[data1['dismissal_kind'] == 'caught and bowled']

# Counting the occurrences of each player being dismissed via caught and bowled
player_dismissals_caught_and_bowled = caught_and_bowled_dismissals['batsman'].value_counts()

# Finding the player dismissed most often via caught and bowled
max_dismissals_player = player_dismissals_caught_and_bowled.idxmax()
max_dismissals_count = player_dismissals_caught_and_bowled.max()

print(f"The player dismissed the maximum number of times via caught and bowled is '{max_dismissals_player}' with {max_dismissals_count} dismissals.")

     
```
